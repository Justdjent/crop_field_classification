```python
import matplotlib.pyplot as plt
from africa_dataset import AfricanRGBDataset
import numpy as np
```

```python
def setup_subplot(sample, i):
    ax = plt.subplot(16, 11, i)
    plt.tight_layout()
    ax.set_title(f'{sample[1]}')
    ax.axis('off')


dataset = AfricanRGBDataset(
    csv_file='data/train_rgb.csv',
    dates=("2017-01-01", "2017-01-31", "2017-02-10", "2017-03-12","2017-03-22","2017-05-31","2017-06-20","2017-07-10","2017-07-15","2017-08-04","2017-08-19"),
    root_dir='./')
fig = plt.figure(figsize=(30,50))

for i in range(len(dataset)):
#     i = i+1
    
    sample = dataset[i]

    print(i, sample[0].shape, sample[1])
    for j, image in enumerate(sample[0]):
        j = j+1
        setup_subplot(sample, 11*i + j)
        image = np.array(image)
        image = np.transpose(image, axes=[1,2,0])
        plt.imshow(image)
#     show_landmarks(**sample)

    if i == 15:
        plt.savefig('hmmm_RGB.pdf')
        plt.show()
        break
```
